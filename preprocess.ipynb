Data Exploration: DEMENTED

```python
Demented = Data[Data.DXCURREN == 1]
Normal = Data[Data.DXCURREN == 0]
```

```python
Demented.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data.plot(kind = 'line', figsize = (12,6), fontsize = 20)
plt.title('Maximum Data Value', fontsize = 20)
plt.grid()
```

```python
import matplotlib.pyplot as plt

x1 = Data['RHIPPOC']
x2 = Demented['RHIPPOC']
x3 = Normal['RHIPPOC']
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs, label='demented')
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)

plt.legend()
plt.show()
```

Checcking for Correlation Between Features

```python
import seaborn as sns
plt.figure(figsize=(22,22))  
ax = sns.heatmap(Data.corr(), annot=True, annot_kws={'size': 20})
col_ax = plt.gcf().axes[-1]
col_ax.tick_params(labelsize=20)
plt.show()
```
